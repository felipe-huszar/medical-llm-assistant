```python
import os, sys

# Configuracao principal do notebook
USE_MOCK = 'false'
BENCHMARK_MODE = 'false'  # true para benchmark isolado (sem historico/persistencia)

MODEL_FILE_ID = '1zPLTPoOmXN5QOYZdN2La99J6iaqsP3sr'
MODEL_ZIP_NAME = 'medical_assistant_llm_14b.zip'
LORA_LOCAL_PATH = '/content/model'

os.environ['USE_MOCK_LLM'] = USE_MOCK
os.environ['BENCHMARK_MODE'] = BENCHMARK_MODE
os.environ['MODEL_PATH'] = LORA_LOCAL_PATH
os.environ['BASE_MODEL_ID'] = 'unsloth/Qwen2.5-14B-Instruct-bnb-4bit'
os.environ['CHROMA_DB_PATH'] = '/content/chroma_db'

sys.path.insert(0, '/content/medical-llm-assistant')

print('=' * 55)
print('CONFIGURACAO ATUAL')
print('=' * 55)
print(f"USE_MOCK_LLM   = {os.environ['USE_MOCK_LLM']}")
print(f"BENCHMARK_MODE = {os.environ['BENCHMARK_MODE']}")
print(f"MODEL_PATH     = {os.environ['MODEL_PATH']}")
print(f"BASE_MODEL_ID  = {os.environ['BASE_MODEL_ID']}")
print(f"MODEL_FILE_ID  = {MODEL_FILE_ID}")
print('=' * 55)

if USE_MOCK.lower() != 'true':
    import subprocess
    import zipfile
    import shutil

    print()
    print('Baixando modelo do Drive (arquivo direto)...')
    os.makedirs(LORA_LOCAL_PATH, exist_ok=True)
    zip_path = f'{LORA_LOCAL_PATH}/{MODEL_ZIP_NAME}'

    result = subprocess.run(
        ['gdown', f'https://drive.google.com/uc?id={MODEL_FILE_ID}', '-O', zip_path],
        capture_output=True, text=True
    )

    if result.returncode != 0 or not os.path.exists(zip_path):
        print('Erro no gdown:', result.stderr)
        print()
        print('Tentando fallback pelo Drive montado...')
        from google.colab import drive
        drive.mount('/content/drive')
        fallback_path = f'/content/drive/MyDrive/FIAP-TC-Fase3/{MODEL_ZIP_NAME}'
        if os.path.exists(fallback_path):
            shutil.copy(fallback_path, zip_path)
            print('ZIP copiado do Drive montado.')
        else:
            raise FileNotFoundError(f'Nao foi possivel baixar o modelo e o fallback nao existe em: {fallback_path}')
    else:
        print('Download concluido.')

    print(f'Extraindo: {MODEL_ZIP_NAME}')
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall(LORA_LOCAL_PATH)
    print('Modelo extraido.')

    cfg = None
    wts = None
    for root, _, files in os.walk(LORA_LOCAL_PATH):
        if 'adapter_config.json' in files and 'adapter_model.safetensors' in files:
            cfg = os.path.join(root, 'adapter_config.json')
            wts = os.path.join(root, 'adapter_model.safetensors')
            break

    if cfg and wts:
        final_model_path = os.path.dirname(cfg)
        os.environ['MODEL_PATH'] = final_model_path
        print('Modelo validado com adapter LoRA.')
        print(f'MODEL_PATH = {final_model_path}')
    else:
        print('Erro: adapter_config.json ou adapter_model.safetensors nao encontrados')
        print(f'Conteudo de {LORA_LOCAL_PATH}:')
        get_ipython().system(f'find {LORA_LOCAL_PATH} -maxdepth 3 -type f | sort')
        raise RuntimeError('Artefato invalido para o loader atual (esperado: adapter-only).')
else:
    print()
    print('Modo Mock ativo - modelo nao necessario.')
```

## 3.5 Benchmark isolado

Use as funcoes abaixo para rodar benchmarks sem reaproveitar historico nem persistir consultas.


```python
import json, time, unicodedata
from pathlib import Path
from src.graph.pipeline import run_consultation
from src.rag.patient_rag import save_patient

def _normalize(text):
    text = unicodedata.normalize("NFKD", text or "").encode("ascii", "ignore").decode("ascii")
    return " ".join(text.lower().strip().split())

def run_benchmark_cases(cases, cpf_prefix="701111333", out_path="/content/benchmark_result.json"):
    results = []
    for idx, case in enumerate(cases, 1):
        cpf = f"{cpf_prefix}{idx:02d}"
        profile = case.get("profile") or {
            "nome": f"Benchmark {idx}",
            "idade": 50,
            "sexo": "F",
            "peso": 70,
            "comorbidades": case.get("comorbidades", []),
        }
        save_patient(cpf, profile)
        started = time.time()
        result = run_consultation(
            cpf=cpf,
            doctor_question=case["question"],
            patient_profile=profile,
            benchmark_mode=True,
        )
        elapsed = time.time() - started
        final_answer = result.get("final_answer", "")
        results.append({
            "id": case.get("id", idx),
            "question": case["question"],
            "expected": case.get("expected"),
            "final_answer": final_answer,
            "elapsed_seconds": round(elapsed, 2),
        })
    payload = {"count": len(results), "results": results}
    Path(out_path).write_text(json.dumps(payload, ensure_ascii=False, indent=2))
    print(f"Benchmark salvo em: {out_path}")
    return payload
```

```python
import sys

# Remove app e src/ do cache — EXCETO src.llm.* que contem o modelo em VRAM
# Apagar src.llm.factory limparia _cached_llm e forcaria reload do modelo
for mod in list(sys.modules):
    if mod == "app" or (mod.startswith("src.") and not mod.startswith("src.llm")):
        del sys.modules[mod]

import app as _app
import src.llm.factory as _factory

# Reutiliza modelo ja em VRAM (cache em src.llm.factory._cached_llm)
if _factory._cached_llm is not None:
    _app.set_llm(_factory._cached_llm)
    print("Modelo reutilizado do cache.")
else:
    print("⚠️ Modelo não está em cache. Rode a célula anterior primeiro.")

_app.demo.queue().queue().launch(share=True)
```
